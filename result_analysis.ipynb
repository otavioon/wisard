```python
from pathlib import Path
import yaml

import pandas as pd
```

```python
experiments = Path("data/wisard-experiments/experiments")
```

```python
results = []

for exp_file in experiments.glob("*.yaml"):
    with exp_file.open("r") as f:
        values = yaml.load(f, Loader=yaml.FullLoader)

    best_bleach = values["best_bleach"]
    accuracy = values["history"]["accuracy"][best_bleach]
    ties = values["history"]["ties"][best_bleach]
    encoder = values["input args"]["encoder"]
    linear_index = values["input args"]["linear_index"]
    resolution = values["input args"]["resolution"]
    tuple_size = values["input args"]["tuple_size"]
    dataset_path = values["input args"]["dataset_path"]
    
    results.append({
        "best_bleach": best_bleach,
        "accuracy": accuracy,
        "ties": ties,
        "clonable": linear_index,
        "resolution": resolution,
        "tuple_size": tuple_size
    })
    
df = pd.DataFrame(results)
    # print(accuracy, ties, encoder, linear_index, resolution, tuple_size)
```
